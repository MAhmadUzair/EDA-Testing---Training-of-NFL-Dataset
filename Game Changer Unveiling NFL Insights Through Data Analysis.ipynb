### Finding Missing Percentages

```python
# Assuming you have a DataFrame called 'df'
missing_values = df.isnull().sum()
total_values = df.shape[0]  # Total number of rows

# Calculate the percentage of missing values
missing_percentage = (missing_values / total_values) * 100

# Create a new DataFrame to display missing values and their percentages
missing_data = pd.DataFrame({'Missing_Values': missing_values, 'Percentage': missing_percentage})

# Print the missing data summary
missing_data
```

```python
missing_percentage = (df.isnull().sum() / len(df)) * 100
```

```python
threshold = 50  # You can adjust this threshold as needed
columns_to_plot = missing_percentage[missing_percentage > threshold].index
```

```python
print("Columns with more than 50% missing values:")
columns_to_plot
```

### PLotting Graphs for Missing percentages

```python
# Plot the missing data percentages
plt.figure(figsize=(24, 6))
plt.bar(missing_data.index, missing_data['Percentage'])
plt.xlabel('Columns')
plt.ylabel('Percentage of Missing Values')
plt.title('Missing Data Percentage per Column')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Plot a bar chart of missing values
plt.figure(figsize=(36, 6))
sns.barplot(x=missing_values.index, y=missing_values.values)
plt.xticks(rotation=90)
plt.xlabel('Columns')
plt.ylabel('Missing Values')
plt.title('Missing Values per Column')
plt.show()
```

### PLotting 5 Highest Missiing Values Columns

```python
# Assuming you have a DataFrame called 'df'
missing_values = df.isnull().sum()
total_values = df.shape[0]  # Total number of rows

# Calculate the percentage of missing values
missing_percentage = (missing_values / total_values) * 100

# Create a new DataFrame to store missing values and percentages
missing_data = pd.DataFrame({'Missing_Values': missing_values, 'Percentage': missing_percentage})

# Sort the DataFrame by 'Percentage' in descending order and select the top 5 columns
top_5_missing = missing_data.sort_values(by='Percentage', ascending=False).head(5)

# Plot the missing data percentages for the top 5 columns
plt.figure(figsize=(10, 6))
bars = plt.bar(top_5_missing.index, top_5_missing['Percentage'])
plt.xlabel('Columns')
plt.ylabel('Percentage of Missing Values')
plt.title('Top 5 Columns with Highest Missing Data Percentage')
plt.xticks(rotation=90, ha='right')  # Rotate x-axis labels to the right by 90 degrees

# Annotate each bar with its percentage value
for bar in bars:
    height = bar.get_height()
    plt.annotate(f'{height:.2f}%', 
                 xy=(bar.get_x() + bar.get_width() / 2, height),
                 xytext=(0, 3),  # 3 points vertical offset
                 textcoords="offset points",
                 ha='center', va='bottom')

plt.show()
```

 Calculate the Total Number of Games Played per Season

```python
games_per_season = df['GameID'].groupby(df['Season']).nunique()

# Create a bar graph to visualize the total number of games per season
plt.figure(figsize=(10, 6))  # Adjust the figure size as needed
games_per_season.plot(kind='bar', color='skyblue')
plt.title('Total Number of Games Played per Season')
plt.xlabel('Season')
plt.ylabel('Total Games Played')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Scatter Plot (Season vs. Yards Gained)

```python
# Create a scatter plot between "season" and "yards_gained"
plt.figure(figsize=(10, 6))
plt.scatter(df['Season'], df['Yards.Gained'], alpha=0.5)
plt.title('Scatter Plot: Season vs. Yards Gained')
plt.xlabel('Season')
plt.ylabel('Yards Gained')
plt.grid(True)
plt.show()
```

Compareing different play types and their yards gained

```python
# Create a box plot to compare different play types and their yards gained
plt.figure(figsize=(12, 6))
sns.boxplot(x='PlayType', y='Yards.Gained', data=df)
plt.title('Box Plot: Play Type vs. Yards Gained')
plt.xlabel('Play Type')
plt.ylabel('Yards Gained')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Play type distribution by down

```python
# Create a bar plot of play type distribution by down
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='down', hue='PlayType')
plt.title('Play Type Distribution by Down')
plt.xlabel('Down')
plt.ylabel('Count')
plt.legend(title='Play Type', loc='upper right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
# Check for null values in the specified target columns
target_columns = [
    "HomeTeam",              
    "AwayTeam",         
    "HomeTeam",      
    "AwayTeam",      
    "PlayType",      
    "Touchdown",      
    "Yards.Gained",     
    "FieldGoalResult",    
    "InterceptionThrown",
    "Sack",           
]

# Create a DataFrame containing only the target columns
target_df = df[target_columns]

# Check for null values in the target DataFrame
null_values_in_target = target_df.isnull().sum()

# Print the null values in the target columns
print("Null values in target columns:")
print(null_values_in_target)
```

```python
for column in target_columns:
    unique_values = df[column].nunique()
    value_counts = df[column].value_counts()
    
    print(f"Column: {column}")
    print(f"Number of unique values: {unique_values}")
    print("Value counts:")
    print(value_counts)
    print("\n")
```
